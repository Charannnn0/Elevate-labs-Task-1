# Rename Columns

```python
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_')
```

# removinng duplicate rows

```python
df = df.drop_duplicates()
```

# Remove Rows with Invalid or Negative Age

```python
df = df[df['age'] >= 0]
```

# Standardize the no_show Column

```python
df['no_show'] = df['no_show'].str.strip().str.capitalize()
```

# Check raw values from the original dataset

```python
df = pd.read_csv('KaggleV2-May-2016.csv')


print(df[['ScheduledDay', 'AppointmentDay']].head(10))
```

# Formatting the column names

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'], errors='coerce')
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'], errors='coerce')
```

# Verify conversion

```python
print(df[['ScheduledDay', 'AppointmentDay']].head())
print(df.dtypes[['ScheduledDay', 'AppointmentDay']])
```

```python
df['WaitDays'] = (df['AppointmentDay'] - df['ScheduledDay']).dt.days
df['Appointment_DayOfWeek'] = df['AppointmentDay'].dt.day_name()
```

```python
df.head()
```

# # Map different values to standard ones

```python
# Convert gender to lowercase
df['Gender'] = df['Gender'].str.lower().str.strip()


df['Gender'] = df['Gender'].replace({
    'f': 'female',
    'm': 'male'
})
```
